## How to create and configure your react native app

### Step 0

check if the path variable and env variable properly configured

### Step 1

npx @react-native-community/cli@latest init MyApp1 --version latest

### Step 2

create a {local.properties} file inside android directory with content {sdk.dir=C:\\Users\\pragy\\AppData\\Local\\Android\\Sdk
}

### Step 3

cd MyApp1

### Step 4

npm install --save-dev @react-native-community/cli@latest

### Step 5

Check if any device is connected or not {adb devices} and also check android tools are present in android studio check for these tools{Open SDK Manager

Open Android Studio.

On the Welcome Screen, click Configure → SDK Manager.

If you are inside a project, it’s File → Settings → Appearance & Behavior → System Settings → Android SDK.

Make sure the following are installed under SDK Tools:

Android SDK Build-Tools

Android Emulator

Android SDK Platform-Tools

Android SDK Tools (sometimes optional)

NDK (Side by Side)

Install any missing ones.}

### Step 5

npx react-native run-android
